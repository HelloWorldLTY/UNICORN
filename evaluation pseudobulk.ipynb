```python
sampled_adata_test = sampled_adata[sampled_adata.obs['enf_set'] == 'test']
pbmc_pred_test = pbmc_pred[pbmc_pred.obs['enf_set'] == 'test']
pbmc_pred_test.X = pbmc_pred_test.layers['predicted']
```

```python
sampled_adata_test.var['train'] = 'train'
```

```python
pbmc_pred_test.var['cell types']
```

```python
# list(sampled_adata_test.obs['gene_symbol'])
```

```python
# sampled_adata_test = sampled_adata_test.T
sampled_adata_test.layers['x_raw'] = sampled_adata_test.X.copy()
sampled_adata_test = pseudo_bulk(sampled_adata_test,
            genes=list(sampled_adata_test.obs['ens_id']),
            cell_type_col='cell types',
            gene_col='ens_id',
            mode='mean',
            expr_threshold=0,
            mem_efficient_mode=True,
            layer='x_raw')
```

```python
# sampled_adata_test = sampled_adata_test.T
pbmc_pred_test = pseudo_bulk(pbmc_pred_test,
            genes=list(pbmc_pred_test.obs['ens_id']),
            cell_type_col='cell types',
            gene_col='ens_id',
            mode='mean',
            expr_threshold=0,
            mem_efficient_mode=True,
            layer='predicted')
```

```python
pbmc_pred_test.X
```

```python
sampled_adata_test.X
```

```python
pbmc_pred_test.X
```

```python
import scipy.stats
import sklearn.metrics
import numpy as np
```

```python
r2_list = []
mse_list = []
for i,j in zip(pbmc_pred_test.X, sampled_adata_test.X):
    accdata = scipy.stats.pearsonr(i,j)[0]
    r2_list.append(accdata)
    mse_list.append(sklearn.metrics.mean_squared_error(i,j))
```

```python
r2_list
print(np.nanmean(r2_list))
print(np.nanstd(r2_list))
```

```python
mse_list
print(np.nanmean(mse_list))
print(np.nanstd(mse_list))
```

```python
import pickle 
dict_data = {}
dict_data['corr'] = r2_list
dict_data['mse'] = mse_list
method = 'seq2cells2'
with open(f"{method}_thymus_hsc_hyenadna_result_ct_hsc_mixtureloss.pickle", "wb") as output_file:
    pickle.dump(dict_data, output_file)
```
